# Plotting a more useful bargraph

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.groupby('sex')['survived'].mean()
```

```python
titanic.groupby('sex')['survived'].mean().plot.bar()
```

```python
titanic.groupby('embark_town').count().loc[:,'survived'].plot.bar()
```

```python
titanic.groupby('embark_town').count().loc[:,'survived']
```
